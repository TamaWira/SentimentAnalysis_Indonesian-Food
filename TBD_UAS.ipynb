```
# Import Library

import re
import json
import pandas as pd
from textblob import TextBlob
from deep_translator import GoogleTranslator
```

```
# Import File
df = pd.read_json('/content/gdrive/My Drive/Makan_use.json')
df
```

```
# Create temporary dataframe

data = pd.DataFrame()

# Append id and tweet from old dataframe to new dataframe
data['id'] = df['_id']
data['location'] = df['Tweet_location']
data['tweet'] = df['Tweet_text']

data
```

```
# Clean the data
def cleanText(text):
  text = re.sub(r'@[a-zA-Z0-9]+', '', text)
  text = re.sub(r',', '', text)
  text = re.sub(r'#', '', text)
  text = re.sub(r'!', '', text)
  text = re.sub(r'&', '', text)
  text = re.sub(r"b'", '', text)
  text = re.sub(r"\'", '', text)
  text = re.sub(r"_.*", '', text)
  text = re.sub(r"wk.*wk", '', text)
  text = re.sub('\[.*?\]', '', text)
  text = re.sub(r'\[a-zA-Z0-9+', '', text)
  text = re.sub(r'\w*\d+\w*', '', text)
  text = re.sub(r'\\.*\\', "", text)
  text = re.sub("https*\S+", " ", text)
  text = re.sub('\s{2,}', " ", text)
  text = text.lower()
  text = text.encode('ascii', 'ignore').decode()

  return text

data['tweet'] = data['tweet'].apply(cleanText)
try:
  print(data)
except Exception as error:
  print(error)
```

```
# Add the label into the DataFrame

kol = ['id', 'tweet']
teks = pd.DataFrame(columns=kol)

Id = []
twit = []
loc = []
label = []

i=0

for text in data['id']:
  tweet = data['tweet']
  if 'seblak' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('seblak')
  elif 'martabak' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('martabak')
  elif 'bakso' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('bakso')
  elif 'rendang' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('rendang')
  elif 'mentai' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('mentai')
  elif 'sushi' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('sushi')
  elif 'rice bowl' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('rice bowl')
  elif 'takoyaki' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('takoyaki')
  elif 'roti bakar' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('roti bakar')
  elif 'sei' in tweet[i]:
    Id.append(data['id'][i])
    twit.append(tweet[i])
    loc.append(data['location'][i])
    label.append('sei')
  
  i=i+1

teks['id'] = Id
teks['tweet'] = twit
teks['location'] = loc
teks['label'] = label

teks
```

```
twit
```

```
# Translate Tweet to English

arti = []

try:
  for i in range(len(twit)):
    arti.append(GoogleTranslator(source='auto', target='en').translate(teks['tweet'][i]))
except Exception as e:
  print(e)

arti
```

```
# Drop polarity and sentiment (if needed)

teks.drop(columns=['polarity', 'sentiment'])
```

```
# Get Polarity
# def getPolarity(text):
#   return TextBlob(text).sentiment.polarity

# Convert Polarity to Sentiment
def showSentiment(polarity):
  if polarity > 0.0:
    return "positif"
  elif polarity == 0.0:
    return "netral"
  else:
    return "negatif"

for i in range(len(arti)):
  teks['polarity'][i] = TextBlob(arti[i]).sentiment.polarity
  # teks['polarity'] = str(artikan[i]).apply(getPolarity)
  # teks['polarity'] = getPolarity(arti)
  teks['sentiment'] = teks['polarity'].apply(showSentiment)

teks
```
